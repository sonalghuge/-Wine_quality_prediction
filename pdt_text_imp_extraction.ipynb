```python
from transformers import pipeline
import PyPDF2

def pdf_to_text(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()

    return text

def post_process_answer(answer, context):
    # Find the start and end of the sentence containing the answer
    start = max(0, context.rfind('.', 0, answer['start']))
    end = min(len(context), context.find('.', answer['end']))

    # Extract the full sentence
    answer_text = context[start:end].strip()
    return answer_text

# Replace 'your_pdf_file.pdf' with the path to your PDF file
pdf_path = '/content/Constitution of India.pdf'
data = pdf_to_text(pdf_path)

# Set up the question-answering pipeline
qa_pipeline = pipeline('question-answering', model='deepset/bert-base-cased-squad2', tokenizer='deepset/bert-base-cased-squad2')

# User input for questions
while True:
    user_question = input("\nAsk a question (or type 'exit' to end): ")

    if user_question.lower() == 'exit':
        break

    # Limit context length
    max_context_length = 4096
    truncated_data = data[:max_context_length]

    # Perform question-answering
    answer = qa_pipeline(question=user_question, context=truncated_data)

    # Post-process the answer to extract the full sentence
    answer_text = post_process_answer(answer, truncated_data)
    print("Answer:", answer_text)
```

```python
from transformers import GPT2Tokenizer, GPT2Model
import torch

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            pdf_document = fitz.open(file)
            text = ''
            for page_num in range(pdf_document.page_count):
                page = pdf_document[page_num]
                text += page.get_text("text")
        return text
    except Exception as e:
        print(f"Error extracting text: {e}")
        return ''

# Specify the path to your PDF file
pdf_file_path = '/content/Constitution of India.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file_path)

# Define a search query
search_query = "10. Power of District Council to make regulations for"

# Combine the search query with the PDF text
combined_text = f"{search_query}\n\n{pdf_text}"

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2")

# Tokenize and convert to PyTorch tensor
inputs = tokenizer(combined_text, return_tensors="pt", max_length=1024, truncation=True)

# Forward pass through the GPT-2 model
with torch.no_grad():
    outputs = model(**inputs)

# Extract embeddings for the search query
query_embedding = outputs.last_hidden_state[0, -1, :].numpy()

# Calculate cosine similarity to find the most similar section
cosine_similarities = torch.nn.functional.cosine_similarity(outputs.last_hidden_state[0, -1, :], torch.tensor(query_embedding), dim=0)
most_similar_index = torch.argmax(cosine_similarities).item()

# Extract the relevant text based on the most similar index
relevant_lines = combined_text.split('\n')[most_similar_index + 1:]
relevant_text = '\n'.join(relevant_lines)[:500]  # Limit to 500 characters

# Print the relevant text
print(relevant_text)
```

```python
from sentence_transformers import SentenceTransformer, util

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            pdf_document = fitz.open(file)
            text = ''
            for page_num in range(pdf_document.page_count):
                page = pdf_document[page_num]
                text += page.get_text("text")
        return text
    except Exception as e:
        print(f"Error extracting text: {e}")
        return ''

# Specify the path to your PDF file
pdf_file_path = '/content/Constitution of India.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file_path)

# Define a search query
search_query = "20. Protection in respect of conviction for offences."

# Load a pre-trained BERT-based model (you can choose a different model)
model = SentenceTransformer('bert-base-uncased')

# Encode the search query and the document text
query_embedding = model.encode(search_query, convert_to_tensor=True)
document_embeddings = model.encode(pdf_text.split('\n'), convert_to_tensor=True)

# Calculate cosine similarities between the query and each sentence in the document
cosine_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)[0]

# Get the index of the most similar sentence
most_similar_index = torch.argmax(cosine_scores).item()

# Extract the relevant text based on the most similar index
relevant_lines = pdf_text.split('\n')[most_similar_index + 1:]
relevant_text = '\n'.join(relevant_lines)[:500]  # Limit to 500 characters

# Print the relevant text
print(relevant_text)
```

```python
from sentence_transformers import SentenceTransformer, util
from concurrent.futures import ProcessPoolExecutor

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            pdf_document = fitz.open(file)
            text = ''
            for page_num in range(pdf_document.page_count):
                page = pdf_document[page_num]
                text += page.get_text("text")
        return text
    except Exception as e:
        print(f"Error extracting text: {e}")
        return ''

def process_batch(sentences, query_embedding):
    document_embeddings = model.encode(sentences, convert_to_tensor=True)
    cosine_scores = util.pytorch_cos_sim(query_embedding, document_embeddings)[0]
    return torch.argmax(cosine_scores).item()

# Specify the path to your PDF file
pdf_file_path = '/content/Constitution of India.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file_path)

# Define a search query
search_query = "20. Protection in respect of conviction for offences."

# Load a pre-trained BERT-based model
model = SentenceTransformer('bert-base-uncased')

# Encode the search query
query_embedding = model.encode(search_query, convert_to_tensor=True)

# Split the document into sentences
sentences = pdf_text.split('\n')

# Define batch size for processing
batch_size = 100

# Use ProcessPoolExecutor for parallel processing
with ProcessPoolExecutor() as executor:
    batches = [sentences[i:i + batch_size] for i in range(0, len(sentences), batch_size)]
    results = list(executor.map(process_batch, batches, [query_embedding] * len(batches)))

# Get the index of the most similar sentence across all batches
most_similar_index = max(results)

# Extract the relevant text based on the most similar index
relevant_lines = pdf_text.split('\n')[most_similar_index + 1:]
relevant_text = '\n'.join(relevant_lines)[:500]  # Limit to 500 characters

# Print the relevant text
print(relevant_text)
```
